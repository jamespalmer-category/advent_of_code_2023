```python
import re
import ast
import numpy as np
import math
from functools import cache
from itertools import product

def read_file_to_string(file_path):
    with open(file_path,'r') as f:
        return f.read()
    
day_13_string = read_file_to_string('day_13_input.txt')
day_13_test = read_file_to_string('day_13_test.txt')
# day_13_test_2 = read_file_to_string('day_13_test_2.txt')
# day_13_test_3 = read_file_to_string('day_13_test_3.txt')
# day_13_test_4 = read_file_to_string('day_13_test_4.txt')
# day_13_test_5 = read_file_to_string('day_13_test_5.txt')
```

```python
def parse_string(string):

    day_13 = string.split('\n\n')
    day_13 = [string.split('\n') for string in day_13]
    day_13 = [np.array([np.array(list(string)) for string in grid]) for grid in day_13]

    return day_13
```

```python
def horizontal_reflection(grid):

    #for each row, check row i refls with either row above or below, 
    
    for i, row in enumerate(grid):
        
        refls_above = np.array_equal(grid[i-1],grid[i]) if i != 0 else False
        #refls_below = np.array_equal(grid[i-1],grid[i]) if i != len(grid)-1 else False

        if refls_above:
            refl_bool = True
            for j in range(min(i, len(grid)-i)):
                #print("index",i)
                #print(grid[i-j-1],grid[i+j])
                #print(not np.array_equal(grid[i-j-1],grid[i+j]))
                if not np.array_equal(grid[i-j-1],grid[i+j]):
                    refl_bool = False


            if refl_bool == True:
                return i*100
            
    return 0
```

```python
def vertical_reflection(grid):
    grid = grid.T
    return horizontal_reflection(grid) // 100
```

```python
def horizontal_reflection_smudge(grid):

    #for each row, check row i refls with either row above or below, 
    
    for i, row in enumerate(grid):
        
        refls_above = (np.array_equal(grid[i-1],grid[i]) or (np.count_nonzero(grid[i-1] != grid[i]) > 1)) if i != 0 else False
        #refls_below = np.array_equal(grid[i-1],grid[i]) if i != len(grid)-1 else False

        if refls_above:
            refl_bool = True
            smudge = False
            for j in range(min(i, len(grid)-i)):
                #print("index",i)
                #print(grid[i-j-1],grid[i+j])
                #print(not np.array_equal(grid[i-j-1],grid[i+j]))
                if not np.array_equal(grid[i-j-1],grid[i+j]):
                    if np.count_nonzero(grid[i-j-1] != grid[i+j]) == 1:
                        if smudge:
                            refl_bool = False
                        else:
                            smudge = True
                    if np.count_nonzero(grid[i-j-1] != grid[i+j]) > 1:
                        refl_bool = False


            if refl_bool and smudge == True:
                return i*100
            
    return 0
```

```python
def vertical_reflection_smudge(grid):
    grid = grid.T
    return horizontal_reflection_smudge(grid) // 100
```
