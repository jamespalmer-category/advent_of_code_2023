```python
import re
import ast
import numpy as np
import math
from functools import cache
from itertools import product

def read_file_to_string(file_path):
    with open(file_path,'r') as f:
        return f.read()
    
day_12_string = read_file_to_string('day_12_input.txt')
day_12_test = read_file_to_string('day_12_test.txt')
# day_12_test_2 = read_file_to_string('day_12_test_2.txt')
# day_12_test_3 = read_file_to_string('day_12_test_3.txt')
# day_12_test_4 = read_file_to_string('day_12_test_4.txt')
# day_12_test_5 = read_file_to_string('day_12_test_5.txt')
```

```python
def parse_string(string):
    day_12 = string.split('\n')
    day_12_seqs = [re.findall(r'[.?#]+', string) for string in day_12]

    #day_12_seqs = [['.']+re.findall(r'[.]+|[?#]+', string)+['.'] for string in day_12]
    day_12_locs = [re.findall(r'[0-9]+', string) for string in day_12]

    return day_12_seqs,[tuple([int(loc) for loc in string]) for string in day_12_locs]
```

```python
@cache
def how_many_combos(string, length):
    if length > len(string):
        return 0
    # string has # or ?s
    count = 0
    indexes = [m.start(0) for m in re.finditer(r'#', string)]
    #print(indexes)
    for i in range(len(string[:-1*length])+1):
        #print(string[i:i+length])
        all_in = True
        for index in indexes:
            #print(index,i+length)
            if index < i or index >= i+length:
                #print("hello")
                all_in = False

        #print(all_in)
        if all_in:
            count += 1
    #print(count)
    return count
```

```python
@cache
def count_combos(string, number_list):
    #print(string, number_list)

    j=0
    min_length = sum(number_list) + len(number_list) - 1
    number_of_combos = 0

    #generalise the below to do this for when number list and number of #s are equal
    blocks = re.findall(r'[#?]+', string)
    blocks = [block for block in blocks if set(block) != {'?'}]
    
    if min_length > len(string):
        return 0
    
    

    elif len(number_list) == len(blocks):

        #print(re.findall(r'\?(?:\#*\?*)*#|#(?:\?*#*)*\?|[#]+', string))
        #print([string[m.start(0): m.end(0)] for m in re.finditer(r'\?(?:\#*\?*)*#|#(?:\?*#*)*\?', string)][0], number_list[0])
        for i, block in enumerate(blocks):#
            #print("index:",i,block, number_list[i])
            if number_of_combos == 0:
                number_of_combos += how_many_combos(block, number_list[i])
                if number_of_combos == 0:
                    return number_of_combos
            else:
                number_of_combos = number_of_combos*how_many_combos(block, number_list[i])
                if number_of_combos == 0:
                    return number_of_combos
        #print("same length:",string,number_of_combos)
        return number_of_combos
        


    
    else:
        #print(string[len(string)-min_length+1:])
        for i in range(len(string[:])):

            if (len(string[i:]) < min_length) or '#' in string[:i]:
                break

            if string[i] == '?':
                # and '#' not in string[i:i+number_list[j]] -- might need this
                if '.' not in string[i:i+number_list[j]]:
                    if i+number_list[j] >= len(string):
                        #print("hello 1")
                        number_of_combos += 1
                        break

                    else:
                        if string[i+number_list[j]] in '.?':
                            if len(number_list) != 1:
                                
                                number_of_combos += 1*count_combos(string[i+number_list[j]+1:], number_list[j+1:])
                                #print("back in:",string)
                                
                            else:
                                #print("hello 2")
                                number_of_combos += 1
                    #break

            elif string[i] == '#':
                #print(i)
                # and '?' not in string[i:i+number_list[j]]
                if '.' not in string[i:i+number_list[j]]:
                    if i+number_list[j] >= len(string):
                        if (string[i-1] != '#') or (i == 0):
                            #print("hello 3")
                            number_of_combos += 1
                            break
                    else:
                        #print(string[i:i+number_list[j]],string[i+number_list[j]])
                        if string[i+number_list[j]] in '.?':
                            if len(number_list) != 1:
                                number_of_combos += 1*count_combos(string[i+number_list[j]+1:], number_list[j+1:])
                                if set(string[i:i+number_list[j]]) == {'#'}:
                                    break
                                
                            else:
                                #print("hello 4")
                                number_of_combos += 1
                                if set(string[i:i+number_list[j]]) == {'#'}:
                                    break
                elif '.' in string[i:i+number_list[j]]:
                    #print("breakout:",string[i:i+number_list[j]], number_of_combos)
                    break
                        
        #if number_of_combos != 0:
            #print("string:",string, "number:",number_of_combos)
        return number_of_combos
```

```python
def part_one(seqs,locs):

    sum_ = 0
    for i,seq in enumerate(seqs):
        #print(count_combos(seq[0], locs[i]))
        sum_ += count_combos(seq[0], locs[i])

    return sum_
```
