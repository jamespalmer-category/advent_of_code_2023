```python
def move_boulders_north(grid):

    grid = grid[::-1]

    for (i,j), element in np.ndenumerate(grid):

        if i != grid.shape[0] -1:

            if element == 'O':

                if grid[i+1][j] == '#':
                    continue
                elif grid[i+1][j] == '.':
                    grid[i][j] = '.'
                    grid[i+1][j] = 'O'

    return grid[::-1]
```

```python
def move_boulders_south(grid):
    return move_boulders_north(grid[::-1])[::-1]
```

```python
def move_boulders_west(grid):
    return move_boulders_north(grid.T).T
```

```python
def move_boulders_east(grid):
    return move_boulders_north(grid.T[::-1])[::-1].T
```

```python
def part_two(grid,num_cycles):
    
    
    sum_list = []

    for i in range(num_cycles):

        old_grid = np.copy(grid)
        new_grid = move_boulders_north(grid)
        while not np.array_equal(new_grid,old_grid):
            #print("hello 1")
            old_grid, new_grid = np.copy(new_grid), move_boulders_north(new_grid)

        #print(np.array_equal(np.where(old_grid=='O'),np.where(new_grid=='O')))
        new_grid = move_boulders_west(new_grid)
        #print(old_grid, new_grid)
        while not np.array_equal(new_grid,old_grid):
            #print("hello 2")
            old_grid, new_grid = np.copy(new_grid), move_boulders_west(new_grid)
        
        new_grid = move_boulders_south(new_grid)
        while not np.array_equal(new_grid,old_grid):
            #print("hello 3")
            old_grid, new_grid = np.copy(new_grid), move_boulders_south(new_grid)

        new_grid = move_boulders_east(new_grid)
        while not np.array_equal(new_grid,old_grid):
            #print("hello 4")
            old_grid, new_grid = np.copy(new_grid), move_boulders_east(new_grid)

        
        sum_ = 0
        for i in range(new_grid.shape[0]):
            sum_ += len(np.where(new_grid[i]=='O')[0]) * (new_grid.shape[0]-i)
        sum_list.append(sum_)

        #build a loop check here that doesn't break when we see only one repeat/two repeats

    return sum_list
```

```python
def find_loop(list, loop_length):
    loop_start = None

    
    for i, element in enumerate(list):
        loop = []
        for j, element_2 in enumerate(list[i:i+loop_length]):
            if i+j+loop_length < len(list):
                if list[i+j+loop_length] == element_2:
                    loop.append(element_2)
                    
            else:
                break
        #print(loop)
        if len(loop) == loop_length:
            loop_start = i
            break

    return loop if len(loop) == loop_length else [], loop_start
```
