```python
import re
import ast
import numpy as np
import math
from functools import cache
from itertools import product

def read_file_to_string(file_path):
    with open(file_path,'r') as f:
        return f.read()
    
day_18_string = read_file_to_string('day_18_input.txt')
day_18_test = read_file_to_string('day_18_test.txt')
#day_18_test_2 = read_file_to_string('day_18_test_2.txt')
# day_18_test_3 = read_file_to_string('day_18_test_3.txt')
# day_18_test_4 = read_file_to_string('day_18_test_4.txt')
# day_18_test_5 = read_file_to_string('day_18_test_5.txt')
```

```python
def parse_string(string):

    dir_dict = {'min_H':0,
                'max_H':0,
                'min_V':0,
                'max_V':0}

    day_18=string.split('\n')

    direction = []
    number = []

    h = 0
    v = 0

    for string in day_18:

        direction.append(string[0])
        number.append(int(re.findall(r'[0-9]+',string)[0]))

        if string[0] == 'U':
            v -= int(re.findall(r'[0-9]+',string)[0])
            dir_dict['min_V'] = min(v,dir_dict['min_V'])  

        elif string[0] == 'D':
            v += int(re.findall(r'[0-9]+',string)[0])
            dir_dict['max_V'] = max(v,dir_dict['max_V'])  

        elif string[0] == 'L':
            h -= int(re.findall(r'[0-9]+',string)[0])
            dir_dict['min_H'] = min(h,dir_dict['min_H'])  

        elif string[0] == 'R':
            h += int(re.findall(r'[0-9]+',string)[0])
            dir_dict['max_H'] = max(h,dir_dict['max_H'])  

    return dir_dict, direction, number
```

```python
def draw_path(dir_dict, instructions, lengths):

    empty_grid = np.zeros((1+dir_dict['max_V'] - dir_dict['min_V'], 1+dir_dict['max_H'] - dir_dict['min_H']))

    position = (-1*dir_dict['min_V'], -1*dir_dict['min_H'])
    x_coords = [-1*dir_dict['min_H']]
    y_coords = [-1*dir_dict['min_V']]

    bdry_points = 0

    for i, instruction in enumerate(instructions):
        #print(i)
        #print(position)


        if instruction == 'U':
            x_coords.append(position[1])
            y_coords.append(position[0]-lengths[i])
            position = (position[0]-lengths[i],position[1])
            

            for j in range(position[0],position[0]+lengths[i]+1):
                bdry_points += 1
                empty_grid[j,position[1]] = 1

            
        elif instruction == 'D':
            x_coords.append(position[1])
            y_coords.append(position[0]+lengths[i])
            position = (position[0]+lengths[i],position[1])
            
            for j in range(position[0]-lengths[i],position[0]+1):
                bdry_points += 1
                empty_grid[j,position[1]] = 1

        elif instruction == 'L':
            x_coords.append(position[1]-lengths[i])
            y_coords.append(position[0]) 
            position = (position[0],position[1]-lengths[i])
            
            for j in range(position[1],position[1]+lengths[i]+1):
                bdry_points += 1
                empty_grid[position[0],j] = 1 

        elif instruction == 'R':
            x_coords.append(position[1]+lengths[i])
            y_coords.append(position[0])
            position = (position[0],position[1]+lengths[i])
            
            #print(position)
            for j in range(position[1]-lengths[i],position[1]+1):
                bdry_points += 1
                empty_grid[position[0],j] = 1


        bdry_points -= 1

    x = np.array(x_coords)
    y = np.array(y_coords)
    area = 1/2 * np.sum(x*np.roll(y, -1) - y*np.roll(x, -1))
    print(abs(area))

    total_area = abs(area) + bdry_points/2 + 1
    print(bdry_points)
        
    return empty_grid, total_area, x, y
```

```python
def parse_string_2(string):

    dir_dict = {'min_H':0,
                'max_H':0,
                'min_V':0,
                'max_V':0}

    day_18=string.split('\n')

    direction = []
    number = []

    h = 0
    v = 0

    for string in day_18:

        if string[-2] == '0':
            direction.append('R')
            string = string[string.index('(')+2:-2]
            number.append(int(string, 16))
            h += int(string, 16)
            dir_dict['max_H'] = max(h,dir_dict['max_H'])  
        elif string[-2] == '1':
            direction.append('D') 
            string = string[string.index('(')+2:-2]
            number.append(int(string, 16))
            v += int(string, 16)
            dir_dict['max_V'] = max(v,dir_dict['max_V']) 
        elif string[-2] == '2':
            direction.append('L') 
            string = string[string.index('(')+2:-2]
            number.append(int(string, 16))
            h -= int(string, 16)
            dir_dict['min_H'] = min(h,dir_dict['min_H']) 
        elif string[-2] == '3':
            direction.append('U') 
            string = string[string.index('(')+2:-2]
            number.append(int(string, 16))
            v -= int(string, 16)
            dir_dict['min_V'] = min(v,dir_dict['min_V'])  


        

    return dir_dict, direction, number
```

```python
def draw_path_2(dir_dict, instructions, lengths):

    position = (0,0)
    x_coords = []
    y_coords = []

    bdry_points = sum(lengths)

    for i, instruction in enumerate(instructions):
        #print(i)
        #print(position)


        if instruction == 'U':
            x_coords.append(position[1])
            y_coords.append(position[0]-lengths[i])
            position = (position[0]-lengths[i],position[1])

            
        elif instruction == 'D':
            x_coords.append(position[1])
            y_coords.append(position[0]+lengths[i])
            position = (position[0]+lengths[i],position[1])

        elif instruction == 'L':
            x_coords.append(position[1]-lengths[i])
            y_coords.append(position[0]) 
            position = (position[0],position[1]-lengths[i])
          

        elif instruction == 'R':
            x_coords.append(position[1]+lengths[i])
            y_coords.append(position[0])
            position = (position[0],position[1]+lengths[i])

    x = np.array(x_coords, dtype=np.int64)
    y = np.array(y_coords, dtype=np.int64)
    area = 1/2 * np.sum(x*np.roll(y, -1) - y*np.roll(x, -1))
    area = abs(area)

    total_area = int(abs(area) + bdry_points/2 + 1)
        
    return total_area
```
